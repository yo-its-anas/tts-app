```python
# Import required libraries
import IPython.display as ipd
from TTS.api import TTS

# Load the TTS model (you can choose a different model if desired)
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC")

# Function to generate speech from text
def generate_speech(text, output_filename='output.wav'):
    # Synthesize speech
    tts.tts_to_file(text=text, file_path=output_filename)
    print(f"Audio saved as {output_filename}")

# Example usage
text_input = "Hello! Welcome to your Text to Speech application."
generate_speech(text_input)

# Play the generated audio
ipd.Audio('output.wav')
```
